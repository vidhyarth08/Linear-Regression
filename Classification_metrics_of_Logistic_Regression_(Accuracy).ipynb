```python
import numpy as np
import pandas as pd
import os

for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
ds = pd.read_csv('iris.data')
```

```python
ds.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(ds.iloc[:,0:-1], ds.iloc[:,-1],test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
clf1 = LogisticRegression()
clf2 = DecisionTreeClassifier()
```

```python
clf1.fit(X_train,y_train)
clf2.fit(X_train,y_train)
```

```python
y_pred1 = clf1.predict(X_test)
y_pred2 = clf2.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print('Accuracy of Logistic Regression:', accuracy_score(y_test,y_pred1))
print('Accuracy of Decision Tree:', accuracy_score(y_test,y_pred2))
```

```python
confusion_matrix(y_test,y_pred1)
```

```python
confusion_matrix(y_test,y_pred2)
```

```python
print("Confusion matrix of Logistic Regression\n")
pd.DataFrame(confusion_matrix(y_test,y_pred1),columns=list(range(0,3)))
```

```python
print("Confusion matrix of Decision Tree\n")
pd.DataFrame(confusion_matrix(y_test,y_pred2),columns=list(range(0,3)))
```

```python
result = pd.DataFrame()
result['Actual Label'] = y_test
result['Logistic Regression Prediction'] = y_pred1
result['Decision Tree Classifier'] = y_pred2
```
