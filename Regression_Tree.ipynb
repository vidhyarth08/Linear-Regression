```python
import pandas as pd
from pandas_datareader import data
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import r2_score
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import GridSearchCV
```

```python
housing = fetch_california_housing()
ds = pd.DataFrame(housing.data, columns=housing.feature_names)
#ds['target'] = housing.target
```

```python
ds['MEDV'] = housing.target
```

```python
ds.head()
```

```python
X = ds.iloc[:,0:8]
y = ds.iloc[:,8]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.2,random_state=42)
```

```python
regt = DecisionTreeRegressor(criterion = 'poisson', max_depth=5)
```

```python
regt.fit(X_train,y_train)
```

```python
y_pred = regt.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
#Hyperparameter tuning
param_grid = {
    'max_depth':[2,5,8,10,None],
    'criterion':['poisson', 'friedman_mse', 'absolute_error', 'squared_error'],
    'max_features':[0.25,0.5,1.0],
    'min_samples_split':[0.25,0.5,1.0],
    'splitter':['best', 'random']
}
```

```python
gs = GridSearchCV(DecisionTreeRegressor(),param_grid=param_grid)
```

```python
gs.fit(X_train,y_train)
```

```python
gs.best_score_
```

```python
gs.best_params_
```

```python
#Feature importance
for importance, name in sorted(zip(regt.feature_importances_, X_train.columns), reverse=True):
    print(name, importance)
```
